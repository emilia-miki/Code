First, let's load the data from the train file:

```python
import pandas as pd
import numpy as np

csv_train = pd.read_csv('train.csv')
csv_train
```

```python
csv_test = pd.read_csv('test.csv')
csv_test
```

```python
for col in csv_train:
    if csv_train[col].dtype == object:
        unique_train = pd.unique(csv_train[col])
        unique_test = pd.unique(csv_test[col])
        if not np.array_equal((unique_train), (unique_test)):
            print('train', unique_train)
            print('test', unique_test)
```

Let's drop the IDs:

```python
csv_train = csv_train.drop('SK_ID_CURR', axis=1)
csv_test = csv_test.drop('SK_ID_CURR', axis=1)
```

Let's write a one-hot encode function for categorical types:

```python
def one_hot_encode(df):
    labels = sorted(pd.unique(df))
    count = len(labels)
    labels_dict = {labels[i]: i for i in range(count)}

    result = np.empty((df.shape[0], count - 1), dtype=np.uint8)
    for i, row in enumerate(df):
        num = labels_dict[row]
        for j in range(count - 1):
            result[i, j] = 1 if j == num else 0

    return result
```

Now let's check if it works:

```python
one_hot_encode(csv_train['NAME_CONTRACT_TYPE'])
```

Let's look at the data in general:

```python
for col in csv_train:
    print(col, pd.unique(csv_train[col]))
```

Let's look at the highest correlations between columns:

```python
for col_i in csv_train:
    if csv_train[col_i].dtype == object:
        continue
    for col_j in csv_train:
        if csv_train[col_j].dtype == object or col_i == col_j:
            continue
        cc = np.corrcoef(csv_train[col_i], csv_train[col_j])[0, 1]
        if cc > 0.5:
            print(col_i, col_j, cc)
```

Let's fill all the nan values in string columns with the string 'XNA':

```python
for col in csv_train:
    if csv_train[col].dtype == object:
        csv_train[col] = csv_train[col].fillna('XNA')
```

Let's check the data again:

```python
for col in csv_train:
    print(col, pd.unique(csv_train[col].sort_values()))
```

The column DAYS_EMPLOYED seems to have an invalid value in it. Let's replace it with NaN.

```python
csv_train['DAYS_EMPLOYED'].value_counts()
```

```python
csv_train['DAYS_EMPLOYED'] = csv_train['DAYS_EMPLOYED'].replace(365243, np.nan)
```

It seems like the AMT_REQ_CREDIT_BUREAU_QRT column has one significant outlier. Let's remove it:

```python
csv_train['AMT_REQ_CREDIT_BUREAU_QRT'].value_counts()
```

```python
csv_train = csv_train[csv_train['AMT_REQ_CREDIT_BUREAU_QRT'] != 261.0]
csv_train = csv_train[csv_train['AMT_REQ_CREDIT_BUREAU_QRT'] != 19.0]
```

Let's move the targets to a separate numpy array:

```python
target = csv_train['TARGET']
target = target.astype(np.uint8)
target = target.to_numpy()
```

And remove them from the original dataset:

```python
csv_train = csv_train.drop('TARGET', axis=1)
```

Let's copy all the columns with categorical values to a separate array and remove them from the dataset:

```python
cols = []
for col in csv_train:
    if csv_train[col].dtype == object:
        cols.append(csv_train[col])
        csv_train = csv_train.drop(col, axis=1)
```

At this point we can convert the dataset to a numpy array:

```python
data = csv_train.to_numpy()
```

```python
data.shape
```

Let's go over all the columns and fill NaN values with the mean of the column, append a new column that indicates whether there was a NaN value in the original dataset, and normalize the data:

```python
data = data.astype(np.float64)
new_cols = np.empty((data.shape[0], 0), np.uint8)
for i in range(data.shape[1]):
    new_col = np.isnan(data[:, i])
    mean = np.nanmean(data[:, i])
    data[:, i] = np.nan_to_num(data[:, i], nan=mean)
    data[:, i] = (data[:, i] - data[:, i].min()) / (data[:, i].max() - data[:, i].min())
    if any(new_col):
        new_col = new_col.astype(np.uint8)
        new_cols = np.concatenate([new_cols, new_col.reshape((new_col.shape[0], 1))], axis=-1)
data = np.concatenate([data, new_cols], axis=-1)
```

Before one-hot encoding the categorical columns, let's compare the encoding to the ones we would get from the test dataset:

```python
len(cols)
```

Now let's one-hot encode all the categorical columns and append them to the data:

```python
for col in cols:
    new_cols = one_hot_encode(col)
    data = np.concatenate((data, new_cols), axis=1)
```

```python
data.shape
```

```python
for i in range(data.shape[1]):
    cc = np.corrcoef(data[:, i], target)[0, 1]
    if cc > 0.05:
        print(i, cc)
```

Now we have two numpy arrays: data and targets, that only consist of normalized float32 values, and we can finally use it to train a model. First, let's construct a simple regression model, train it, and look at the results:

```python
import tensorflow as tf

strategy = tf.distribute.MirroredStrategy()

with strategy.scope():
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(296,)),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(),
                loss=tf.keras.losses.BinaryFocalCrossentropy(),
                metrics=tf.keras.metrics.AUC())
```

```python
np.unique(target, return_counts=True)
```

```python
model.fit(data, target, batch_size=64, epochs=100, class_weight={0: 0.9, 1: 0.1},
          callbacks=[tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5)])
```
